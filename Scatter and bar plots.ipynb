## Making figures with numpy arrays

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
#create some data
x=np.linspace(0,10,100)
x[:10]
```

```python
#plot the data
fig, ax=plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax=plt.subplots()
ax.scatter(x, np.exp(x));
```

## Sine wave

```python
fig, ax=plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
fig, ax=plt.subplots()
ax.scatter(x, np.cos(x));
```

## Making plot from dictionary

```python
items={ "Orange":30,
       "Apple" : 20,
       "Grapes": 25,
       "Mango" :40, 
      "Watemelon" :50}
fig, ax=plt.subplots()
ax.bar(items.keys(), items.values()) 
ax.set(title="Items", ylabel="Price");
```
