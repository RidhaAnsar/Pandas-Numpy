 ## Demo of  standard deviation and variance

```python
import numpy as np
high=np.array([100,200,200, 300])
low=np.array([2,4,6,8])
np.var(high), np.var(low)
```

```python
np.std(high), np.std(low)
```

```python
np.mean(high), np.mean(low)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high)
plt.show()
```

```python
plt.hist(low)
plt.show()
```
