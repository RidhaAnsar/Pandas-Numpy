## Barh, Histogram and subplots

```python
%matplotlib inline
import matplotlib.pyplot as plt
items={"Books":20, 
       "Pen":10,
       "Scale":15,
       "Eraser":7,
       "Pencil":8}

fig, ax= plt.subplots()
ax.barh(list(items.keys()), list(items.values()));
ax.set(xlabel="Price" );
```

```python
import numpy as np
x=np.random.rand(1000)
fig, ax=plt.subplots()
ax.hist(x);
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randn(1000)
fig,((ax1, ax2), (ax3, ax4))=plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
items={"Books":20, 
       "Pen":10,
       "Scale":15,
       "Eraser":7,
       "Pencil":8}
ax3.bar(items.keys(), items.values());

ax4.hist(np.random.rand(100));

```
