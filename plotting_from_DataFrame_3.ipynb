```python
import numpy as np
import pandas as pd
x=np.random.rand(10,4)
x
#turn into dataframe
df=pd.DataFrame(x, columns=['a','b','c', 'd'])
df
```

```python
df.plot.bar();
```

```python
df.plot.hist();
```

```python

car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="bar")
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="hist", bins=20)
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="hist")
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="pie")
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="scatter")
```
