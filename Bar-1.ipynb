```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create bar graph
plt.bar(categories, values, color='skyblue')

# Add title and labels
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```
